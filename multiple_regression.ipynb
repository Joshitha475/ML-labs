```python
# Step 1: Import necessary libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Step 2: Create synthetic data
# Generating random data
np.random.seed(0)
n = 100
X1 = np.random.rand(n) * 10  # Independent variable 1
X2 = np.random.rand(n) * 5   # Independent variable 2
# Dependent variable with some noise
Y = 3 + 2 * X1 + 1.5 * X2 + np.random.randn(n)

# Step 3: Create a DataFrame
data = pd.DataFrame({'X1': X1, 'X2': X2, 'Y': Y})

# Step 4: Define independent and dependent variables
X = data[['X1', 'X2']]  # Independent variables
Y = data['Y']            # Dependent variable

# Step 5: Add a constant to the model (for the intercept)
X = sm.add_constant(X)

# Step 6: Fit the multiple regression model
model = sm.OLS(Y, X).fit()

# Step 7: Display the regression results
print(model.summary())

# Step 8: Plot the actual vs predicted values
predicted_values = model.predict(X)
plt.scatter(Y, predicted_values)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--')  # Line of equality
plt.show()
```
